##### LEETCODE 705. Design HashSet  (easy) [only elements,no key-value pair]
##### LEETCODE 706. Design HashMap (easy) [key-value pair]

https://leetcode.com/problems/design-hashset/description/
https://leetcode.com/problems/design-hashmap/description/
<br><br>
<div>

</div>

```python
#solution

class MyHashSet:
    def __init__(self):
        self.m = {}

    def add(self, key: int) -> None:
        self.m[key] = True

    def remove(self, key: int) -> None:
        if key in self.m:
            del self.m[key]

    def contains(self, key: int) -> bool:
        return key in self.m

hashs=MyHashSet()
hashs.add(1)
hashs.add(2)
print(hashs.m)
print(hashs.contains(1))
hashs.remove(1)
hashs.m
```

```python
#solution

class MyHashMap:
    def __init__(self):
        self.m = {}

    def add(self, key: int,value) -> None:
        self.m[key] = value

    def remove(self, key: int) -> None:
        if key in self.m:
            del self.m[key]

    def get(self, key: int) -> bool:
        if key in self.m:
            return self.m[key]

hashm=MyHashMap()
hashm.add(1,100)
hashm.add(2,200)
hashm.add(4,300)
print(hashm.m)
print(hashm.get(1))
hashm.remove(1)
hashm.m
```

```python
# Print the minimum of all the elements and its key in the heap.

#key
#min() return the value in the first value in sorted.
#key designate the way to sort the values. key=dict.get means the list will be sorted by values of the dictionary.
print(min(hashm.m, key=hashm.m.get))

#value
print(min(hashm.m.values()))
```

###### LEETCODE 2283. Check if Number Has Equal Digit Count and Digit Value (EASY)
https://leetcode.com/problems/check-if-number-has-equal-digit-count-and-digit-value/

<br>
<div>

</div>

```python
#Approach: Create a dict storing frequency of each number and then just compare the index with frequency.

from collections import Counter

def digitCount(num: str) -> bool:
    d = Counter(num)
    print(d)
    for i in range(len(num)):
        if int(num[i])!=d.get(str(i)):
            return False
    return True

print(digitCount("1210"))
print(digitCount("030"))
```

```python
def colorful(A):
    Astring=str(A)
    if len(Astring)<2 & A!=1:
        return 0
    res = [int(a) * int(b) for a, b in zip(Astring, Astring[1:])]
    for i in Astring:
        res.append(int(i))
    c=Counter(res)
    for i in c.values():
        if i!=1:
            return 0
    return 1
                  
colorful(2) #0
colorful(1) #1
colorful(123) #0
```


##### LEETCODE 1. Two Sum [easy]
https://leetcode.com/problems/two-sum/description/ <br>
https://www.interviewbit.com/problems/2-sum/

<br><br>
<div>

</div>

```python
def twoSum(nums, target):
    seen = {}
    for i, value in enumerate(nums):
        remaining = target - value
        ###In each iteration check if required number (required  number = target sum - current number) is
        ###present in the hashmap.

        ###If present, return {required number index, current number index} as  result.

        ###Otherwise add the current iteration number as key and its index as value to the hashmap.
        ###Repeat this  until you find the result.

        if remaining in seen:
            # print(i,value,remaining)
            return [min(i,nums.index(remaining))+1, max(i,seen[remaining])+1] #4,8 (earliest)
#             return [i+1, seen[remaining]+1]  # [8,6]  latest

        seen[value] = i
    return []

twoSum([ 4, 7, -4, 2, 2, 2, 3, -5, -3, 9, -4, 9, -7, 7, -1, 9, 9, 4, 1, -4, -2, 3, -3, -5, 4, -7, 7, 9, -4, 4, -8 ],-3)
```

##### LEETCODE 36. Valid Sudoku (medium)

https://leetcode.com/problems/valid-sudoku/

<br>
<div>

<div>

```python
def isValidSudoku(board) -> bool:
	s = set()

	for i in range(9):
		for j in range(9):
			if board[i][j] != '.':
				c = board[i][j]

				# Row check
				key = str(c) + ' in row ' + str(i)

				if key in s:
					return False
				else:
					s.add(key)

				# Column check
				key =  str(c) + ' in col ' + str(j)

				if key in s:
					return False
				else:
					s.add(key)

				# Box check
				boxIndex = (i // 3) * 3 + (j // 3)
				key = str(c) + ' in box ' + str(boxIndex)

				if key in s:
					return False
				else:
					s.add(key) 

	return True

isValidSudoku([["8","3",".",".","7",".",".",".","."]
,["6",".",".","1","9","5",".",".","."]
,[".","9","8",".",".",".",".","6","."]
,["8",".",".",".","6",".",".",".","3"]
,["4",".",".","8",".","3",".",".","1"]
,["7",".",".",".","2",".",".",".","6"]
,[".","6",".",".",".",".","2","8","."]
,[".",".",".","4","1","9",".",".","5"]
,[".",".",".",".","8",".",".","7","9"]])
```

# pattern: Key formation
#####  ANAGRAM
https://www.interviewbit.com/problems/anagrams/ [medium]
<br><br>
<div>

</div>

```python
from collections import defaultdict

def anagrams(A):
    d = defaultdict(list)
    for i in range(len(A)):
        d["".join(sorted(A[i]))].append(i+1)
    return [v for k,v in d.items()]

anagrams(['dog','gdo'])
```

##### Points on the straight line
https://www.interviewbit.com/problems/points-on-the-straight-line/ [medium] <br>

```python
# Concept: A set of points are on a line each of their pair-wise slopes has the same value.

# For each point, we iterate through all the other points and find the slope with each of the other points and store the number of pairs which have the same slope and find the max of the number of points.

# Time Complexity: O(N^2)
# Space Complexity: O(N) as we track only N-1 lines in total.

def maxPoints(points) -> int:
    if len(points) <= 2:
        return len(points)

    def find_slope(p1, p2):
        x1, y1 = p1
        x2, y2 = p2
        if x1-x2 == 0:
            return math.inf
        return (y1-y2)/(x1-x2)

    ans = 1
    for i, p1 in enumerate(points):
        slopes = defaultdict(int)
        for j, p2 in enumerate(points[i+1:]):
            slope = find_slope(p1, p2)
            slopes[slope] += 1
            ans = max(slopes[slope], ans)
    return ans+1

maxPoints([[1,1],[3,2],[5,3],[4,1],[2,3],[1,4]]) #4
```
