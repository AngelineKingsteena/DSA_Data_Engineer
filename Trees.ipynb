```python
# Implementation of BST in Python:
# Python program to demonstrate insert operation in binary search tree 

#Defining each node containing some value and its children
class TreeNode: 
    def __init__(self,key): 
        self.left = None
        self.right = None
        self.data = key 

# Function to insert node in tree recursively
def insertNode(root,node): 
    if root is None: 
        root = node 
    else: 
        if root.data < node.data: 
            if root.right is None: 
                root.right = node 
            else: 
                insertNode(root.right, node) 
        else: 
            if root.left is None: 
                root.left = node 
            else: 
                insertNode(root.left, node) 

# Function to print inorder traversal recursively
def inOrderTraversal(root): 
    if root: 
        inOrderTraversal(root.left) 
        print(root.data) 
        inOrderTraversal(root.right) 



# Creating a new BST with root as 50
r = TreeNode(55) 
insertNode(r,TreeNode(35)) 
insertNode(r,TreeNode(25)) 
insertNode(r,TreeNode(45)) 
insertNode(r,TreeNode(75)) 
insertNode(r,TreeNode(65)) 
insertNode(r,TreeNode(85)) 
# Print inoder traversal of the BST 
inOrderTraversal(r)
```

## INORDER TRAVERSAL (LEFT-> ROOT -> RIGHT)

```python
from typing import List
######################################################

#   Approach I : Using Recursion

#   Runtime: 24ms   -   95.86%
#   Memory: 14MB  -   92.19%

######################################################

def iterate(root): 
    inorder=[]
    def inOrderTraversal2(root):
        if root: 
            inOrderTraversal2(root.left) 
            inorder.append(root.data) 
            inOrderTraversal2(root.right)
#         return inorder
    ans=inOrderTraversal2(root)
    return inorder
print("from recurion ",iterate(r))


######################################################

#   Approach II : Without Recursion

#   Runtime: 24ms   -   95.86%
#   Memory: 14MB  -   92.19%

## LEFT-->ROOT--> RIGHT
def inorderTraversalroot(root: TreeNode) -> List[int]:
    inorder = []
    stack = []
    #start from node
    node = root
    while stack or node:
        while node:
            stack.append(node)
            node = node.left
        node = stack.pop()
        inorder.append(node.data)
        node = node.right
    return inorder
print("from stack approch ",inorderTraversalroot(r))
```

## PRE-ORDER TRAVERSAL ( ROOT -> LEFT -> RIGHT)

```python
from typing import List
######################################################

#   Approach I : Using Recursion
#   Runtime: 24ms   -   95.86%
#   Memory: 14MB  -   92.19%

######################################################

def recursive2(root): 
    preorder=[]
    def preorder_recursive(root):
        if root: 
            preorder.append(root.data)
            preorder_recursive(root.left)
            preorder_recursive(root.right) 
    preorder_recursive(root)
    return preorder
print("from recurion ",recursive2(r))


######################################################

#   Approach II : Without Recursion
#   Runtime: 24ms   -   95.86%
#   Memory: 14MB  -   92.19%

# ROOT -> LEFT -> RIGHT
def preorder_stack(root: TreeNode) -> List[int]:
    preorder = []
    stack = []
    #start from node
    node = root
    stack.append(node)
    while stack :
        node = stack.pop()
        preorder.append(node.data)
        if node.right:
            stack.append(node.right)
        if node.left:
            stack.append(node.left)   
    return preorder
print("from stack approch ",preorder_stack(r))
```

## Post-ORDER TRAVERSAL ( LEFT -> RIGHT -> ROOT)

```python
from typing import List
######################################################

#   Approach I : Using Recursion
#   Runtime: 24ms   -   95.86%
#   Memory: 14MB  -   92.19%

######################################################

def recursive3(root): 
    postorder=[]
    def preorder_recursive(root):
        if root:
            preorder_recursive(root.left)
            preorder_recursive(root.right) 
            postorder.append(root.data)
            
    preorder_recursive(root)
    return postorder
print("from recurion ",recursive3(r))


######################################################

#   Approach II : Without Recursion
#   Runtime: 24ms   -   95.86%
#   Memory: 14MB  -   92.19%

#  LEFT -> RIGHT -> ROOT
def preorder_stack(root: TreeNode) -> List[int]:
    post_order, stack = [], []
    prev, p = None, root
    while stack or p:
        if p:
            stack.append(p)
            p = p.left
        else:
            top = stack[-1]
            if top.right and top.right != prev:
                p = top.right
            else:
                post_order.append(top.data)
                prev = top
                stack.pop()
    return post_order
print("from stack approch ",preorder_stack(r))
```

```python
#iterative
def printLevelOrder(root):
    # Base Case
    ans=[]
    if root is None:
        return
 
    # Create an empty queue
    # for level order traversal
    queue = []
 
    # Enqueue Root and initialize height
    queue.append(root)
 
    while(len(queue) > 0):
 
        # Print front of queue and
        # remove it from queue
        ans.append(queue[0].data)
#         print(queue[0].data, end = " ")
        node = queue.pop(0)
 
        # Enqueue left child
        if node.left is not None:
            queue.append(node.left)
 
        # Enqueue right child
        if node.right is not None:
            queue.append(node.right)
    return ans
printLevelOrder(r)
```
