```python
#expalantion https://www.youtube.com/watch?v=v-Ta5GyxRxc&ab_channel=Innoskrit
import random

def solve(nums, k):
    pivot= random.choice(nums)
    left = [x for x in nums if x > pivot]
    mid = [x for x in nums if x == pivot]
    right= [x for x in nums if x <pivot]
    L,M=len(left),len(mid)
    if k <= L:
        return solve(left,k)
    elif k > (L+M):
        return solve(right,k-(L+M))
    else:
#         print(nums,left,right,mid)
        return mid[0]

solve([11, 3, 4, 6],3) #[4,6,11]
```
