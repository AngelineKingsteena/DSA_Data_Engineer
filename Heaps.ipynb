```python
import heapq

H = [21,1,45,78,3,5]
# Covert to a heap
heapq.heapify(H)
print(H)

# Add element
heapq.heappush(H,8)
heapq.heappush(H,8)
print(H)

# Remove element from the heap
heapq.heappop(H)
print(H)

# Replace an element
#The heap replace function always removes the smallest element of the heap and inserts
#the new incoming element at some place not fixed by any order.

heapq.heapreplace(H,9) #3(smallest) replaced by 9
print(H)
heapq.heapreplace(H,2) #5(smallest) replaced by 2
print(H)
```

#####  JESSE AND COOKIES [easy]

```python
#explanation https://www.youtube.com/watch?v=cJtq8eZCWpk&ab_channel=HackersRealm
from heapq import heapify, heappop, heappush

def cookies(k, A):
    heapify(A)
    count = 0
    # main logic
    while True:
        x = heappop(A)
        #base condition
        if x >=k:
            return count
        if A:
            y = heappop(A)
            s = x + 2 * y
            heappush(A, s)
            count += 1
        else:
            return -1

cookies(70,[1,7,57,48])
```

#####  LEETCODE 23.Merge k Sorted Lists/ [hard]

```python
from typing import List,Optional
from heapq import heapify, heappop, heappush

# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next

def mergeKLists(lists: List[Optional[ListNode]]) -> Optional[ListNode]:
    h =[]
    for i,l in enumerate(lists):
        if l:heappush(h,(l.val,i))
    dummy = curr = ListNode(0)
    while h:
        val, i = heappop(h)
        curr.next=ListNode(val)
        if lists[i].next:
            heappush(h,(lists[i].next.val,i))
            lists[i]= lists[i].next
        curr=curr.next
    return dummy.next

mergeKLists([[1,4,5],[1,3,4],[2,6]])
```

## MAX HEAP & MIN HEAP

### pattern 
### use min heapsort for largest,farthest,longest,etc
### use max heapsort for smallest,nearest,shortest,etc

### TIME COPLEXITY
## finding min and max in min heap and max heap is 0(1) respectively
## adding or removing elements from heap is 0(logn)

 explanation: https://www.thinkxacademy.com/Data%20Structures%20in%20Python/ <br>
 https://www.thinkxacademy.com/Data%20Structures%20in%20Python/ 

```python
##this is max heap sort
def max_heapify(Arr, n, i):
    largest = i
    left = 2*i + 1
    right = 2*i + 2

    #largest so far is compared with left child
    if left < n and Arr[largest] < Arr[left]:
        largest = left


    #largest so far is compared with right child
    if right < n and Arr[largest] < Arr[right]:
        largest = right
        
    #change parent
    if largest!=i:
        Arr[i], Arr[largest] = Arr[largest], Arr[i]
        #recursive call
        max_heapify(Arr, n, largest)
    
def max_heapsort(Arr):
    n = len(Arr)
    # Build a max heap
    for i in range(n//2 - 1, -1, -1):
        max_heapify(Arr, n, i)
    return Arr

    

#Driver Code
Arr = [2,66,30,5,9,10]

max_heapsort(Arr)
```

```python
import random

def solve(nums, k):
        min_heap =[]
        heapq.heapify(min_heap)
        for i in range(k):
            heapq.heappush(min_heap, nums[i])
        for i in range(k, len (nums)):
            if nums[i] > min_heap[0]:
                heapq.heappop(min_heap)
                heapq.heappush(min_heap, nums[i])
        return list(min_heap)

solve([11, 3, 4, 6],3) #[4,6,11]
```

#####  INTERVIEWBIT Merge K sorted arrays! [medium]

```python
from heapq import heapify, heappop, heappush,merge

def solve(lists):
    #    return list(merge(*lists))
    li = lists[0]
    heapify(li)
    for elem in lists[1:]:
        for x in elem:
            heappush(li,x)
    lis = []
    while li:
        lis.append(heappop(li))
    return lis

solve([ [1, 2, 3],
        [2, 4, 6],
        [0, 9, 10]
     ])
```

```python
def solve(A, B, C):
    minHeap = []
    heapLen = 0
    for a in A:
        for b in B:
            heapq.heappush(minHeap, a+b)
            heapLen += 1
            if heapLen > C:
                heapq.heappop(minHeap)
                heapLen -= 1
    res = []
    while minHeap:
        res.append(heapq.heappop(minHeap))
    return res[::-1]
# Time: O(N^2 log(C))
# Space: O(C)

solve([1, 4, 2, 3],[2, 5, 1, 6],4)
```

#####  LEETCODE  295. Find Median from Data Stream [hard]

```python
#explanation https://www.youtube.com/watch?v=itmhHWaHupI&ab_channel=NeetCode

from heapq import heapify,heappush,heappop,merge
class MedianFinder:
    def __init__(self):
        #max heap
        self.small=[]
        #min heap
        self.big=[]
    def addNum(self, num: int) -> None:
        #since default heapify implementation in min heap
        heappush(self.small,-num)
        # heapify(self.small) #tle
        # heapify(self.big) #tle 
        if ((self.small and self.big) and (-self.small[0]> self.big[0])):
            val=-heappop(self.small)
            heappush(self.big,val)
        #uneven sizes on either partitions
        if (len(self.small)>len(self.big)+1):
            n=-heappop(self.small)
            heappush(self.big,n)
        if (len(self.big)>len(self.small)+1):
            n=heappop(self.big)
            heappush(self.small,-n)
           

    def findMedian(self) -> float:
        if len(self.small)>len(self.big):
            return -self.small[0]
        elif len(self.small)<len(self.big):
            return self.big[0]
        else:
            return (-self.small[0]+self.big[0])/2
    
medianFinder = MedianFinder()
medianFinder.addNum(1)    #// arr = [1]
medianFinder.addNum(2)    #// arr = [1, 2]
medianFinder.findMedian() #// return 1.5 (i.e., (1 + 2) / 2)
medianFinder.addNum(3)    #// arr[1, 2, 3]
medianFinder.findMedian()
```

```python
import heapq
from typing import List
from collections import Counter
def topKFrequent(nums: List[int], k: int) -> List[int]:
    c=Counter(nums)
    #for max heap implementation
    c = [(-v,k)for k,v in c.items()]
#     print(c)
    heapq.heapify(c)
#     print(c)
    output=[]
    for i in range(k):
        item= heapq.heappop(c)
        output.append(item[1])
    return output

topKFrequent([1,2,3,1,2,1],2) #[1,2] 
```
